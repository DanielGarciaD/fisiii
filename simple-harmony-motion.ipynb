```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
k = 1.0
m = 1.0

pi = 3.1416

A = 1.0
phi = 0.0
ti = 0.0
tf = 13

#w = math.sqrt(k/m)
w = 1
```

```python
def position (w0, t0, A0, phi0):
    x = A0*math.cos(w0*t0+phi0)
    return x;

def velocity (w0, t0, A0, phi0):
    v = -A0*w0*math.sin(w0*t0+phi0)
    return v;

def acceleration (w0, t0, A0, phi0):
    a = -A0*w0*w0*math.cos(w0*t0+phi0)
    return a;
```

```python
t = np.arange(ti,tf,0.1)
pos = []
vel = []
acc = []

for i in t:
    pos.append(position(w,i,A,phi))
    vel.append(velocity(w,i,A,phi))
    acc.append(acceleration(w,i,A,phi))

fig, ax = plt.subplots()

ax.plot(t,pos,'r--',label="Position")
ax.plot(t,vel,'b-*',label="Velocity")
plt.plot(t,acc,'g^',label="Acceleration")

#start, end = ax.get_xlim()
start = 0.0
end = tf
ax.xaxis.set_ticks(np.arange(start, end, 0.5))

plt.grid(True)
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.rcParams['figure.figsize'] = (12.0, 4.0)

plt.show()
```
